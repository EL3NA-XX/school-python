3. Schrijf een functie somDelers(n)
die de som berekent van alle delers van n behalve n zelf.
Gebruik deze functie om te bepalen of het getal perfect is (bv. 6 = 1 + 2 + 3).

```python
def somDelers(n):
    som = 0
    for i in range(1, n):
        if n % i == 0:
            som += i
    return som


n = int(input("Geef een getal: "))

if somDelers(n) == n:
    print("perfect getal")
else:
    print("geen perfect getal")
```

4. Schrijf een functie isMeetkundig(lijst) die onderzoekt of een reeks getallen een meetkundige rij vormt.
Een meetkundige rij is een lijst getallen waarbij elk getal (vanaf het tweede) ontstaat door het vorige getal te vermenigvuldigen met een vaste verhouding (ratio).
Voorbeeld: bij verhouding 3 hoort de rij
2, 6, 18, 54, …

De functie moet het volgende doen:

-Controle op lengte
Als de lijst minder dan 3 getallen bevat, kan je niet bepalen of er sprake is van een meetkundige rij.
In dat geval geeft de functie False terug.

-Bepalen van de verhouding
Voor de eerste twee getallen bereken je de verhouding:
verhouding = tweede getal / eerste getal
Deze verhouding mag een kommagetal zijn.

-Controle op meetkundige rij
Voor elk volgend getal controleer je of:
huidig getal / vorig getal gelijk is aan dezelfde verhouding.

Als dit ergens niet klopt, is het geen meetkundige rij.

Return-waarde
Als de lijst lang genoeg is, geeft de functie altijd een tuple terug in de vorm:
(True of False, verhouding)

True betekent: de rij is meetkundig.

False betekent: de rij is niet meetkundig.

De verhouding moet ook worden teruggegeven, zelfs als de rij niet klopt.

```python
def isMeetkundig(lijst):
    if len(lijst) < 3:
        return False, None
    
    rat = lijst[1] / lijst[0]
    for i in range(1, len(lijst)):
        if lijst[i] / lijst[i-1] != rat:
            return False, rat
    return True, rat

print(isMeetkundig([2, 6, 18, 54]))
print(isMeetkundig([2, 6, 19, 54]))
print(isMeetkundig([5, 10]))
```

8. Deck genereren 

Genereer een volledige deck als lijst van tuples, bijv. ('Harten', 7).
Print daarna:

het totaal aantal kaarten

alle kaarten waarvan de waarde groter dan 8 is (10, Boer, Dame, Heer, Aas)

```python
kleuren = ['harten(rood)', 'ruiten(rood)', 'schoppen(zwart)', 'klaver(zwart)']
waarden = [2, 3, 4, 5, 6, 7, 8, 9, 10, 'Boer', 'Dame', 'Heer', 'Aas']

deck = []
for kleur in kleuren:
    for waarde in waarden:
        deck.append((kleur, waarde))

print("Volledige deck:", len(deck))

hkaarten = []
for kaart in deck: 
    if kaart[1] == 'Boer' or kaart[1] == 'Dame' or kaart[1] == 'Heer' or kaart[1] == 'Aas' or kaart[1] > 8:
        hkaarten.append(kaart)

print("Kaarten met waarde groter dan 8:", hkaarten)
```

9. Letterfrequentie 
Bepaal voor het woord:
woord = 'microbiologischeprocessen'


hoe vaak elke letter voorkomt

welke letter het meest voorkomt

hoeveel verschillende letters er zijn

```python
woord = 'microbiologischeprocessen'
f = {}

for letter in woord: 
    if letter in f: 
        f[letter] += 1
    else:
        f[letter] = 1 

meest = ''
hoogste = 0
for letter in f:
    if f[letter] > hoogste:
        hoogste = f[letter]
        meest = letter

print("Meest voorkomende letter:", meest)
```

11. Schrijf een functie insertion_sort(lijst)
Implementeer insertion sort in Python. Gebruik alleen basisoperaties; geen ingebouwde sorteermethoden.

Maak testdata
Genereer lijsten met willekeurige getallen van verschillende groottes, bijvoorbeeld:

100 elementen

200 elementen

400 elementen

800 elementen

Meet de uitvoeringstijd
Gebruik time.perf_counter() om te meten hoe lang insertion sort erover doet.
Voor elke lijstgrootte voer je het algoritme minstens 3 keer uit en neem je de gemiddelde tijd.

Verwerk de resultaten
Maak een tabel waarin je zet:
'lijstgrootte' en   'gemeten tijd'
Zet de waarden vervolgens in een excel grafiek.

Analyse
Onderzoek hoe de tijd verandert wanneer de lijst twee keer zo groot wordt.

Wordt de tijd ongeveer twee keer zo groot?

Vier keer zo groot?

Of anders?

Conclusie
Leid uit je metingen zelf af wat de tijdscomplexiteit van insertion sort is in het slechtste geval.
Beschrijf dit in één of twee zinnen.




Doe hetzelfde maar dan voor selection sort

```python
import random
import time

# insertion sort

def insertion_sort(lst):
    for i in range(1, len(lst)):
        k = lst[i]
        j = i - 1
        while j >= 0 and lst[j] > k:
            lst[j + 1] = lst[j]
            j -= 1
        lst[j + 1] = k

groottes = [100, 200, 400, 800]
rondes = 3

print("Insertion Sort timing:")
for size in groottes:
    td = []
    for _ in range(rondes):
        testlijst = [random.randint(0, 1000) for _ in range(size)]
        start = time.perf_counter()
        insertion_sort(testlijst)
        einde = time.perf_counter()
        td.append(einde - start)
    gemiddelde = sum(td) / rondes
    print("Lijstgrootte: " + str(size) + ", Gemiddelde tijd: " + str(gemiddelde) + " sec")

# selection sort

def selection_sort(lst):
    n = len(lst)
    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            if lst[j] < lst[min_index]:
                min_index = j
        lst[i], lst[min_index] = lst[min_index], lst[i]


print("Selection Sort timing:")
for size in groottes:
    td = []
    for _ in range(rondes):
        testlijst = [random.randint(0, 1000) for _ in range(size)]
        start = time.perf_counter()
        selection_sort(testlijst)
        einde = time.perf_counter()
        td.append(einde - start)
    gemiddelde = sum(td) / rondes
    print("Lijstgrootte: " + str(size) + ", Gemiddelde tijd: " + str(gemiddelde) + " sec")
```
